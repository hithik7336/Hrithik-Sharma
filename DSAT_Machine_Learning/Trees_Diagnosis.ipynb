```python
X = data.iloc[:,:30]
```

```python
y = data.iloc[:,30]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score
```

```python
print('The Confusion Matrix Is: ','\n','\n',confusion_matrix(y_test, y_pred))
print('\n')
print('The Accuracy Score Is: ',accuracy_score(y_test, y_pred))
print('\n')
print('The Recall Score Is: ',recall_score(y_test, y_pred))
print('\n')
print('The Precision Score Is: ',precision_score(y_test, y_pred))
print('\n')
print('The F1 Score Is: ',f1_score(y_test,y_pred))
print('\n')
print('The Classification Report Is: ','\n','\n',classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(confusion_matrix(y_pred, y_test), annot = True, cmap = 'autumn', linewidths= 10, linecolor='black')
```
